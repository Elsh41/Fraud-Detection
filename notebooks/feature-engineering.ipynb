```python
# ==========================================
# 4. TRANSFORMATION & IMBALANCE HANDLING (Train Only)
# ==========================================

import pandas as pd
import numpy as np
import sys
sys.path.append('../')

# 1. Load BOTH cleaned datasets directly from disk
try:
    features_fraud = pd.read_csv('../data/processed/intermediate_features_fraud.csv')
    credit_df = pd.read_csv('../data/processed/cleaned_creditcard.csv')
    print("Successfully loaded both datasets into memory!")
except FileNotFoundError as e:
    print(f"Error: Missing processed file. Make sure you fully ran both EDA notebooks first.\n{e}")


# 2. Proceed with your imports and modeling prep
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE

# --- TRANSFORM FRAUD DATA ---
X_f = features_fraud.drop(columns=['class'])
y_f = features_fraud['class']

X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_f, y_f, test_size=0.2, random_state=42, stratify=y_f)

num_cols_f = ['purchase_value', 'age', 'hour_of_day', 'day_of_week', 'time_since_signup', 'device_velocity', 'ip_velocity']
cat_cols_f = ['source', 'browser', 'sex', 'country']

preprocessor_f = ColumnTransformer([
    ('num', StandardScaler(), num_cols_f),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols_f)
])

X_train_f_scaled = preprocessor_f.fit_transform(X_train_f)
X_test_f_scaled = preprocessor_f.transform(X_test_f)

# Apply SMOTE strictly to train set
smote = SMOTE(random_state=42)
X_train_f_res, y_train_f_res = smote.fit_resample(X_train_f_scaled, y_train_f)

# Save processed variables
np.save('../data/processed/X_train_f.npy', X_train_f_res)
np.save('../data/processed/y_train_f.npy', y_train_f_res)
np.save('../data/processed/X_test_f.npy', X_test_f_scaled)
np.save('../data/processed/y_test_f.npy', y_test_f)

# --- TRANSFORM CREDIT CARD DATA ---
X_c = credit_df.drop(columns=['Class'])
y_c = credit_df['Class']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_c, y_c, test_size=0.2, random_state=42, stratify=y_c)

scaler = StandardScaler()
# Note: creditcard dataset is completely numerical except 'Time' and 'Amount'
cols_to_scale = ['Time', 'Amount'] + [c for c in X_c.columns if c.startswith('V')]
X_train_c_scaled = scaler.fit_transform(X_train_c[cols_to_scale])
X_test_c_scaled = scaler.transform(X_test_c[cols_to_scale])

X_train_c_res, y_train_c_res = smote.fit_resample(X_train_c_scaled, y_train_c)

# Save processed variables
np.save('../data/processed/X_train_c.npy', X_train_c_res)
np.save('../data/processed/y_train_c.npy', y_train_c_res)
np.save('../data/processed/X_test_c.npy', X_test_c_scaled)
np.save('../data/processed/y_test_c.npy', y_test_c)
```
